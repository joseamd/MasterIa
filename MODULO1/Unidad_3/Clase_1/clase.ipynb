## No todos los nulos son iguales

### TIPO QUÉ SIGNFICA
MCAR:    Falta complementar al azar, cuando el nulo no tiene relación con otras variables, ni con el valor que falta.  
MAR:     Relacionado con otra variable,   
MNAR:    Relacionado con el valor que falta  

¿Con qué valor reemplazamos el hueco?

### ESTRATEGIA  CUÁNDO USARLA  
Media:      Numérica, distribución simétrica, sin outliers.  mean()  
Mediana:    Numérica con outliers o distribución sesgada. median()     
Moda:       Variable categórica. mode()  
Por grupo:  El nulo depende de otra variable  

```python
# Imputación 1 Salario ---> Mediana

# Aquí se toma la columna salario_cop del DataFrame df y se calcula su mediana.
mediana_salario = df['salario_cop'].median()
print(f"Mediana de salario: ${mediana_salario:,.0f}")

# fillna() reemplaza todos los valores vacíos o nulos (NaN) de la columna salario_cop por el valor de la mediana.
df['salario_cop'] = df['salario_cop'].fillna(mediana_salario)
```

```python
# Imputación 2 ---> Nivel Estres, Media
media_estres = df['nivel_estres'].mean()
print(f"Media de Nivel Estres: ${media_estres}")

df['nivel_estres'] = df['nivel_estres'].fillna(media_estres)
```

```python
# Imputación 3, Satisfación laboral
mediana_laboral = df['satisfaccion_laboral'].median()
print(f"Mediana de Satisfación laboral: ${mediana_laboral}")

df['satisfaccion_laboral'] = df['satisfaccion_laboral'].fillna(mediana_laboral)
```

```python
# Imputación 4, ciudad con moda

# Aquí el código toma la columna ciudad del DataFrame df y calcula su moda. [0] se toma el primer valor de esa serie.
ciudad_moda = df['ciudad'].mode()[0]
print(f'Moda de la ciudad: {ciudad_moda}')

# Esta línea reemplaza todos los valores vacíos (NaN) de la columna ciudad por la moda calculada.
df['ciudad'] = df['ciudad'].fillna(ciudad_moda)
```

```python
# Se valida de nuevo el DataFrame y ya no debe tener datos nulos
df.isnull().sum()
```

## Cuartiles

EL IQR es la distancia entre Q1 y Q3
$$
IQR = Q3 - Q1
$$
$$
Limite\;superior = Q1 - (1.5*IQR)
$$
$$
Limite\;inferior = Q3 + (1.5*IQR)
$$

```python
Q1 = df['horas_semana'].quantile(0.25)
Q3 = df['horas_semana'].quantile(0.75)
df.

IQR = Q3 - Q1

print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
```

```python
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

print(f"Límite inferior: {limite_inferior:.2f} horas")
print(f"Límite superior: {limite_superior:.2f} horas")
```

```python
# Identificar outliers a través de los límites inferior y superior

outliers_iqr = df[
    (df['horas_semana'] < limite_inferior) |
    (df['horas_semana'] > limite_superior)
]

print(f"Outliers detectados en horas_semana: {len(outliers_iqr)}")
print(outliers_iqr[['id_trabajador', 'cargo', 'nivel', 'horas_semana']])
```
