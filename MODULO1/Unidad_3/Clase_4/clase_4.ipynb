```python
import pandas as pd
import numpy as np

df = pd.read_csv('streaming_latam.csv')
df.head(10)
```

```python
print(df.dtypes)
```

```python
print("Paises: ", df['pais'].unique())
print("Géneros de contenido: ", df['genero_contenido'].unique())
print("Dispositivos: ", df['dispositivo'].unique())
```

```python
print(df['cancelo_suscripcion'].value_counts())
print(f"Tasa de cancelación: {df['cancelo_suscripcion'].mean():.1%}")
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['dispositivo_label'] = le.fit_transform(df['dispositivo'])

mapeo = dict(zip(le.classes_, le.transform(le.classes_)))
print("Mapeo:", mapeo)
df[['dispositivo', 'dispositivo_label']].drop_duplicates().sort_values('dispositivo_label')
```

```python
dispositivo_ohe = pd.get_dummies(df['dispositivo'], prefix='dispositivo')
dispositivo_ohe.head(8)
```

```python
dispositivo_ohe_sin_redundancia = pd.get_dummies(df['dispositivo'], prefix='dispositivo', drop_first=True)
dispositivo_ohe_sin_redundancia.head(8)
```

```python
columnas_ohe = ['pais', 'dispositivo', 'genero_contenido', 'comparte_cuenta']

df_encoded = pd.get_dummies(df, columns=columnas_ohe, drop_first=True)

print(f"Columnas en df original: {df.shape[1]}")
print(f"Columnas en df_encoded: {df_encoded.shape[1]}")
```

```python
columnas_nuevas = [c for c in df_encoded.columns if c not in df.columns]
print("Columnas nuevas generadas por OHE:")
print(columnas_nuevas)
```

```python
df['horas_por_titulo'] = df['horas_semana'] / (df['titulos_semana'] + 0.5)
df[['horas_semana', 'titulos_semana', 'horas_por_titulo', 'cancelo_suscripcion']].head(10)
```

```python
df['volumen_actividad'] = df['horas_semana'] * df['titulos_semana']

print("Volumen promedio — NO canceló:")
print(df[df['cancelo_suscripcion'] == 0]['volumen_actividad'].mean().round(2))
print("\nVolumen promedio — SÍ canceló:")
print(df[df['cancelo_suscripcion'] == 1]['volumen_actividad'].mean().round(2))
```
