## CONTROL DE VERSIONES

### 1. ¿Qué es Git y GitHub?

**Git**
- Sistema de control de versiones.
- Vive en tu computador (local).
- Se usa desde terminal o interfaz gráfica.

**GitHub**
- Plataforma web en la nube.
- Permite guardar repositorios Git en internet.
- Requiere una cuenta.

### Instalación de Git

1. Descargar Git desde la página oficial según el sistema operativo.
2. Instalar normalmente.
3. Abrir **Git Bash**.

### Configuración inicial

```python
git config --global user.name "Nombre"
git config --global user.email "el-mismo-email-de-tu-cuenta-github@ejemplo.com"
git config --global init.defaultBranch main  #Configura por defecto la rama main
git config --list
```

### COMANDOS BÁSICOS DE TERMINAL:

```python
mkdir   #Crea una carpeta ej: mkdir nombre_carpeta
cd      #Me permite acceder a una carpeta ej: cd nombre_carpeta
ls -a   #Me permite ver todos los archivos inclusive los ocultos en la carpeta que esté ubicado
```

### COMANDOS BÁSICOS DE GIT:

```python
git init        #Incializa un nuevo repositorio Git local en mi proyecto actual
git ignore      #Archivo que me permite añadir los nombres de los archivos a ignorar
                #Para crearlo, ejecuto el comando: code .gitignore
git status      #Me permite ver que archivos nuevos o modificados están pendiente para el próximo commit
git add .       #Añade cambios de tu directorio de trabajo al área de preparación (staging area o índice
git add .gitignore  #Añade los cambios especificos de un archivo
git commit -m "Agregamos el archivo gitignore"   #Registra los cambios localmente, lo que permite un seguimiento detallado y control de la evolución del proyecto
git log         #Muestra el historial de commits del repositorio
```

### CONECTAMOS CON GITHUB

Una vez creado el repositorio en GitHub, conectamos el repositorio local:

```python
git remote  add origin https://github.com/usuario/nombre_repositorio.git 
git push -u origin main     #Subimos el proyecto por primera vez
git push                    #Subir cambios posteriores 
```

### MANEJO DE RAMAS

```python
git branch      #Muestra el listado de las ramas existentes
git checkout -b nombre_nueva_rama #Comando antiguo, crea rama nombre_nueva_rama localmente
git switch -c nombre_nueva_rama   #Comando moderno, crea rama nombre_nueva_rama localmente
git switch nombre_nueva_rama      #Cambiar de rama
git push -u origin nombre_nueva_rama
```

### FUSIONAR RAMAS

```python
git swith main   #Primero nos ubicamos en main
git merge nombre_nueva_rama #Fusionamos la rama
git push    #Finalmente subimos los cambios
```
