## Bloque B — IA, Machine Learning y Deep Learning

**¿Son lo mismo?**

El director de ShopSmart dijo *"quiero IA"*. Pero IA, Machine Learning y Deep Learning no son sinónimos — son tres herramientas distintas para problemas distintos. Elegir la incorrecta tiene consecuencias reales en tiempo, costo y resultado.

Piénsalos como tres círculos concéntricos:

```
┌──────────────────────────────────────┐
│  INTELIGENCIA ARTIFICIAL             │
│  Todo sistema que hace algo que      │
│  antes solo hacía un humano          │
│                                      │
│   ┌──────────────────────────────┐   │
│   │  MACHINE LEARNING            │   │
│   │  La máquina aprende de datos │   │
│   │  sin que nadie le escriba    │   │
│   │  las reglas                  │   │
│   │                              │   │
│   │   ┌──────────────────────┐   │   │
│   │   │   DEEP LEARNING      │   │   │
│   │   │   Redes neuronales   │   │   │
│   │   │   con muchas capas   │   │   │
│   │   └──────────────────────┘   │   │
│   └──────────────────────────────┘   │
└──────────────────────────────────────┘
```

| Concepto | En palabras simples |
|----------|--------------------|
| **IA** | Cualquier programa que hace algo que antes solo hacían los humanos |
| **Machine Learning** | Le das ejemplos históricos → el sistema descubre las reglas solo |
| **Deep Learning** | ML con redes neuronales de múltiples capas. Necesita muchos datos. |

**La diferencia clave:**
- Programación tradicional → tú escribes las reglas, la máquina las ejecuta
- Machine Learning → tú das ejemplos, la máquina descubre las reglas

**Para ShopSmart:** predecir si un cliente vuelve es un problema con datos en tabla → **ML clásico**.
Si analizáramos fotos de los productos → Deep Learning. Lo verás en el Módulo 2.

| Problema | Herramienta |
|----------|-------------|
| Predecir retorno de clientes (datos en tabla) | ML clásico (scikit-learn) |
| Clasificar imágenes de productos | Deep Learning |
| Analizar reseñas de clientes en texto | Deep Learning (NLP) |
| Reglas de negocio fijas que no cambian | Programación tradicional |

````python
# ── Celda 2 — Sin ML: el programador inventa las reglas ─────────────────────
def predecir_cliente(dia_sin_compra, num_compras, gasto_total):
    if dia_sin_compra < 30 and num_compras > 3:
        return "Definitivamente va a volver"
    elif gasto_total > 500:
        return "Puede volver"
    else:
        return "❌no piensa volver"
    
print(predecir_cliente(17, 1, 200.99))
print(predecir_cliente(29, 8, 1700.99))
print(predecir_cliente(328, 4, 65.28))
````

````python
# ── Celda 3 — Con ML: el modelo aprende las reglas de datos históricos ───────
from sklearn.ensemble import RandomForestClassifier
import numpy as np

X_train = np.array([
    [10, 8, 450],
    [5, 12, 850],
    [20, 6, 50],
    [95, 1, 150],
    [120, 2, 50],
    [60, 8, 280],
    [15, 5, 350],
    [80, 8, 100]
])
y_train = np.array([1, 1, 1, 0, 0, 0, 1, 0])

modelo = RandomForestClassifier(n_estimators=10, random_state=42)
modelo.fit(X_train, y_train)

clientes_nuevos = np.array([
    [17, 1, 200.99],
    [29, 8, 1700.99],
    [328, 4, 65.28]
])

predicciones = modelo.predict(clientes_nuevos)

for i, predicion in enumerate(predicciones):
    estado = "Va a volver" if predicion == 1 else "No va a volver"
    print(f"{[i]}: {estado}")
````

## Bloque C — Ética y Sesgos en IA

> *Un modelo aprende lo que los datos dicen — incluyendo sus errores históricos.*

El modelo de la Celda 3 aprendió de ejemplos pasados. Eso es poderoso. Pero tiene un riesgo que pocas veces se menciona: **si los datos históricos tienen sesgos, el modelo los aprende, los amplifica y los convierte en decisiones automáticas a escala.**

**Tres casos reales:**

| Caso | Qué pasó | Por qué |
|------|----------|--------|
| Amazon CVs (2018) | La IA penalizaba la palabra "mujeres" en los CVs | Fue entrenada con historial de contrataciones donde dominaban los hombres |
| Reconocimiento facial MIT (2018) | 0.8% de error en hombres de piel clara vs 34.7% en mujeres de piel oscura | Los datos de entrenamiento no representaban a todos los grupos |
| Apple Card (2019) | Discriminación por código postal | El código postal correlaciona con raza — el modelo discriminó sin que nadie lo programara |

**En ShopSmart:** si clientes de ciertas zonas compran menos, ¿es porque no quieren comprar? ¿O porque los costos de envío son más altos para esas zonas? Si el modelo aprende ese patrón y dejamos de enviarles ofertas, perpetuamos una inequidad con cada predicción automática.

**Los 3 principios — en cada proyecto, antes de hacer deploy:**

1. **Representatividad** → ¿mis datos cubren a todos los afectados?
2. **Transparencia** → ¿puedo explicar cada decisión del modelo?
3. **Impacto diferencial** → si el modelo se equivoca, ¿quién paga el precio?

````python
# ── Celda 4 — Los 3 principios éticos aplicados a ShopSmart ─────────────────

principios_eticos = {

    "1. REPRESENTATIVIDAD": {
        "pregunta": "¿Los datos representan a TODOS los tipos de clientes?",
        "en_shopSmart": "¿Tenemos datos de clientes nuevos, de zonas rurales, de distintos niveles de gasto?",
        "riesgo": "El modelo funcionará bien para el grupo mayoritario y mal para el resto."
    },

    "2. TRANSPARENCIA": {
        "pregunta": "¿Podemos explicar por qué el modelo tomó cada decisión?",
        "en_shopSmart": "Si un cliente pregunta por qué no recibió una oferta, ¿qué le decimos?",
        "riesgo": "Decisiones automatizadas que nadie puede justificar generan desconfianza y riesgo legal."
    },

    "3. IMPACTO DIFERENCIAL": {
        "pregunta": "Si el modelo se equivoca, ¿el error afecta más a algún grupo específico?",
        "en_shopSmart": "Un falso negativo (no contactar a alguien que iba a volver) ¿perjudica más a ciertos clientes?",
        "riesgo": "El costo del error no se distribuye igual entre todos."
    }

}

for principio, contenido in principios_eticos.items():

    print(f"\n{'─' * 52}")
    print(f"  {principio}")
    print(f"{'─' * 52}")
    print(f"  Pregunta  → {contenido['pregunta']}")
    print(f"  ShopSmart → {contenido['en_shopSmart']}")
    print(f"  Riesgo    → {contenido['riesgo']}")

````

## Bloque D — CRISP-DM: el proceso estándar

**El problema sin proceso:**

El director nos pide el modelo. ¿Por dónde empezamos? ¿Abrimos el CSV y corremos un algoritmo? Sin un orden claro, el 90% de los proyectos de datos falla — no por falta de habilidad técnica, sino por falta de proceso.

**CRISP-DM** *(pronunciado "CRISP Di Em")* es el proceso estándar que usa la industria. Son seis pasos en ciclo — se puede volver a pasos anteriores cuando los resultados lo requieren.

```
  1. Business Understanding  →  ¿Qué necesita el negocio exactamente?
            ↓
  2. Data Understanding      →  ¿Qué datos tenemos? ¿Qué problemas tienen?
            ↓
  3. Data Preparation        →  Limpiar y transformar para el modelo
            ↓
  4. Modeling           ←───   Entrenar y comparar modelos
            ↓        (volver si los resultados no son buenos)
  5. Evaluation              →  ¿El modelo resuelve el problema real?
            ↓
  6. Deployment              →  Poner en producción
```

| Paso | En ShopSmart | Cuándo |
|------|-------------|--------|
| 1. Business Understanding | Definir qué significa "cliente en riesgo" | **Hoy** |
| 2. Data Understanding | Explorar y diagnosticar el CSV | **Hoy** |
| 3. Data Preparation | Limpiar nulos, fechas, encoding | Clase 2 |
| 4. Modeling | Entrenar Random Forest, comparar modelos | Módulo 2 |
| 5. Evaluation | Precisión, Recall, F1-Score | Módulo 2 |
| 6. Deployment | API que genera lista de riesgo cada lunes | Módulo 3 |

> *"El CRISP-DM no es burocracia. Es la diferencia entre construir modelos útiles y construir modelos que nadie usa."*

## Diagnóstico de calidad

````python
import pandas as pd

df = pd.read_csv('clientes_shopSmart.csv')

print("DIAGNÓSTICO DE CALIDAD — ShopSmart")
print("=" * 50)

# --- 1. Valores nulos ---
# .isnull() marca con True cada celda vacía del DataFrame.
# .sum() cuenta los True de cada columna (True vale 1 en Python).
print("\nValores nulos por columna:")
nulos = df.isnull().sum()
for col, n in nulos.items():
    # .items() recorre el resultado columna por columna, dando el nombre y el conteo
    estado = f"{n} nulos — hay que tratar esto antes de modelar" if n > 0 else "Sin nulos"
    print(f"   {col:25s}: {estado}")

# --- 2. Formatos de fecha ---
# .str.match() aplica un patrón de texto a cada celda de la columna.
# El patrón r'^\d{1,2}/' detecta fechas que empiezan con 1 o 2 dígitos seguidos de '/'.
# Eso indica formato DD/MM (como '2/12/2023') en lugar del estándar YYYY-MM-DD.
fechas_problema = df['fecha_ultima_compra'].str.match(r'^\d{1,2}/').sum()
print(f"\nFechas en formato DD/MM/YYYY (inconsistentes) : {fechas_problema} ")
print(f"   Fechas en formato YYYY-MM-DD (estándar)      : {len(df) - fechas_problema}")

# --- 3. Desbalance en la variable objetivo ---
# 'volvio' es lo que el modelo intentará predecir.
# .value_counts() cuenta cuántas veces aparece cada valor único (0 o 1).
conteo = df['volvio'].value_counts()
total  = len(df)
print(f"\nDistribución de 'volvio' (variable a predecir):")
print(f"Sí volvieron (1): {conteo.get(1,0):3d} clientes  ({conteo.get(1,0)/total*100:.1f}%)  ← mayoría")
print(f"No volvieron (0): {conteo.get(0,0):3d} clientes  ({conteo.get(0,0)/total*100:.1f}%)")
print()
````

## Bloque E — Primer contacto con NumPy

**La pregunta que justifica NumPy:**

El director quiere que el modelo analice a sus clientes. Ya cargamos el CSV y vimos sus datos. Pero hay un problema: el CSV tiene texto (`"Ropa"`), fechas (`"2/12/2023"`), identificadores (`"CLI-0001"`). Un modelo de ML no puede procesar ninguna de esas cosas directamente.

**Todo modelo de IA, sin excepción, trabaja con matrices de números.**

Texto, imágenes, fechas, categorías — todo se convierte a números antes de entrar al modelo. NumPy es la librería que hace esa conversión posible y eficiente.

**¿Por qué no usar listas de Python?**

```python
# Lista Python: multiplicar por 2 DUPLICA la lista
[100, 200, 300] * 2  →  [100, 200, 300, 100, 200, 300]

# NumPy: multiplicar por 2 actúa sobre CADA elemento
np.array([100, 200, 300]) * 2  →  [200, 400, 600]
```

NumPy es entre **50 y 200 veces más rápido** para operaciones matemáticas masivas. Con 200 clientes la diferencia es pequeña. Con miles de iteraciones de entrenamiento sobre el dataset completo, la diferencia se convierte en horas.

**Las tres estructuras que usaremos:**

| Estructura | Dimensiones | Pregunta que responde |
|-----------|-------------|----------------------|
| **Vector** | 1D — lista de números | ¿Cómo represento a **un** cliente? |
| **Matriz** | 2D — tabla de vectores | ¿Cómo represento a **todos** los clientes? |
| **Tensor** | 3D+ — matrices apiladas | ¿Cómo representaría imágenes de productos? |

````python
# ── Celda 7 — VECTOR: un cliente representado como números ──────────────────
import numpy as np

# np.array() convierte una lista de Python en un array de NumPy.
# A diferencia de una lista, solo acepta un tipo de dato — eso lo hace eficiente.
# Columnas: [dias_sin_compra, num_compras, gasto_total, num_categorias, cancelaciones]
cli_0002 = np.array([17, 1, 200.99, 2, 0])
cli_0008 = np.array([29, 8, 1646.28, 3, 3])
cli_0001 = np.array([328, 4, 64.65, 3, 1])

print("Representación de clientes como vectores:")
print(f"  CLI-0002: {cli_0002}")
print(f"  CLI-0008: {cli_0008}")
print(f"  CLI-0001: {cli_0001}")
print()

# .shape devuelve las dimensiones del array como tupla.
# (5,) significa: 1 dimensión, 5 elementos.
print(f"Dimensiones de un cliente: {cli_0002.shape}")
print(f"  → 5 características, 1 dimensión = esto es un VECTOR")
print()

# Los índices empiezan en 0. cli_0002[0] es el primer elemento.
print(f"CLI-0002 lleva {int(cli_0002[0])} días sin comprar")
print(f"CLI-0001 lleva {int(cli_0001[0])} días sin comprar — señal de alarma")

````

````python
# ── Celda 8 — MATRIZ: todos los clientes representados como números ──────────

import pandas as pd
import numpy as np

df = pd.read_csv('clientes_shopSmart.csv')

# Tomamos solo las columnas numéricas por ahora.
# Las columnas de texto (categoria_favorita) y fechas necesitan
# transformación antes de poder convertirse en números-
columnas_numericas = ['dias_sin_compra', 'num_compras', 'gasto_total',
                      'num_categorias', 'cancelaciones']

# .values convierte el DataFrame de Pandas a un array de NumPy.
# Este es el "puente" entre exploración de datos y el modelo de ML.
X = df[columnas_numericas].values  # las características de entrada: qué sabemos del cliente
y = df['volvio'].values             # lo que queremos predecir: ¿volvió o no?

# .shape en una matriz devuelve (filas, columnas)
print(f"X — características de entrada: {X.shape}")
print(f"  → {X.shape[0]} clientes (una fila por cliente)")
print(f"  → {X.shape[1]} características (una columna por variable)")
print()
print(f"y — variable a predecir: {y.shape}")
print(f"  → {y.shape[0]} etiquetas, una por cliente")
print()

# X[:3] selecciona las primeras 3 filas.
# El : solo significa "desde el inicio hasta", seguido del límite.
print("Primeras 3 filas de X (los números que verá el modelo):")
print(X[:3])
print()
print("Sus etiquetas en y (lo que el modelo aprenderá a predecir):")
print(y[:3])
````
