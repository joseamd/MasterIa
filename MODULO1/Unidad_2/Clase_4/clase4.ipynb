```python
import sqlite3
conexion = sqlite3.connect("datajobs.db")
```

```python
query_salarios = """
    SELECT 
        candidato_id, 
        tecnologia, 
        salario_actual 
    FROM historial_salarios 
    WHERE salario_actual > 3000000
    ORDER BY candidato_id DESC

"""

df_salarios = pd.read_sql(query_salarios, conexion)
df_salarios.head()
df_salarios.info()
```

## Limpieza

```python
#print(df_candidatos.isna().head())
print(df_candidatos.isna().sum())

total_filas = len(df_candidatos)

porct_nulos = (df_candidatos.isna().sum()/ total_filas) *100
print(porct_nulos.round(2))
```

```python
# Rellenando los nulos

tc_comun = df_candidatos["tecnologia_principal"].mode()[0]
print(tc_comun)
```

```python
df_candidatosNuevo = df_candidatos.copy()
```

```python
df_candidatosNuevo["tecnologia_principal"].fillna(tc_comun)
```
