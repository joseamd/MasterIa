### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí

# Promedio
suma = 0.0
for esperanza in esperanzas_de_vida:
    suma += esperanza
promedio = suma / len(esperanzas_de_vida)
print(f"El promedio de la lista es: {promedio:.1f}")

# Máximo
mayor = esperanzas_de_vida[0] # Inicializamos con el primer elemento
for esperanza in esperanzas_de_vida:
    if mayor <= esperanza:
        mayor = esperanza
    else:
        mayor = mayor
print(f"El número mayor de la lista es: {mayor:.1f}")

# Mínimo
minimo = esperanzas_de_vida[0] # Inicializamos con el primer elemento
for esperanza in esperanzas_de_vida:
    if minimo >= esperanza:
        minimo = esperanza
    else:
        minimo = minimo
print(f"El número mínimo de la lista es: {minimo:.1f}")
````

### 1.3 — Funciones

Las funciones nos permiten reutilizar lógica.

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

````python
# Tu código aquí
def clasificar_esperanza_vida(age):
    if age >= 75:
        return "Alta"
    elif  age >= 60:
        return "Media"
    else:
        return "Baja"

print(clasificar_esperanza_vida(-6))
print(clasificar_esperanza_vida(20))
print(clasificar_esperanza_vida(66))
print(clasificar_esperanza_vida(74.9))
print(clasificar_esperanza_vida(80))
````

### 2.2 — Limpieza de datos

🔧 **Ejercicio 2.2** — El dataset tiene algunos valores nulos (¡datos reales siempre tienen este problema!).  
1. Rellena los valores nulos de columnas **numéricas** con la **media** de cada columna
2. Verifica que ya no queden nulos
3. Explica en una celda de texto: ¿Por qué usamos la media y no simplemente eliminamos esas filas?

````python
# Tu código aquí
# 1. Rellenar nulos con la media de cada columna numérica
salud_limpio = salud.fillna(salud.mean(numeric_only=True))
print("-" * 60)

# 2. Verificar que no queden nulos
print(salud_limpio.isnull().sum())
print("-" * 60)
````

**Responde aquí** (doble clic para editar):

*¿Por qué usamos la media para rellenar nulos en vez de eliminar las filas?*

> Tu respuesta:  
Porque eliminar filas es como tirar todo el expediente de un paciente solo porque le falta un dato. Puede que ese país tenga información valiosa en otras columnas y se perdería por completo.
En cambio, rellenar con la media es como decir "no sé el valor exacto, pero pongo algo razonable basado en el resto de los datos". No es perfecto, pero es mucho mejor que perder información útil, especialmente cuando el dataset no es muy grande.

### 2.3 — Estadísticas descriptivas

🔧 **Ejercicio 2.3**
1. Muestra el resumen estadístico completo del dataset
2. ¿Cuál es la esperanza de vida **promedio** mundial?
3. ¿Cuál es el país con **mayor** gasto en salud? ¿Y el de **menor**?
4. ¿Cuál es la **mediana** de la mortalidad infantil?

````python
# Tu código aquí
# 1. Resumen estadístico
salud_limpio.describe()
print("-" * 60)

# 2. Esperanza de vida promedio mundial
promedio = salud_limpio["esperanza_vida"].mean()
print(f"El promedio mundial es: {promedio:.1f}")
print("-" * 60)

# 3. País con mayor y menor gasto en salud
mayor_gasto = salud_limpio.loc[salud_limpio["gasto_salud_usd"].idxmax(), "pais"]
menor_gasto = salud_limpio.loc[salud_limpio["gasto_salud_usd"].idxmin(), "pais"]
print(f"El paías con mayor gasto es: {mayor_gasto}")
print(f"El país con menor gasto: {menor_gasto}")
print("-" * 60)

# 4. La mediana de la mortalidad infantil
mediana = salud_limpio["mortalidad_infantil"].median()
print(f"La mediana de mortalidad infantil: {mediana:.1f}")
print("-" * 60)
````

### 2.4 — Filtrado y selección

🔧 **Ejercicio 2.4**
1. Filtra los países de **América** (`region == "Americas"`) y muestra solo las columnas: `pais`, `esperanza_vida`, `gasto_salud_usd`
2. Filtra los países con esperanza de vida **mayor a 80 años** — ¿cuántos son? ¿De qué regiones son?
3. Filtra los países con **nivel_ingresos == "Low"** y calcula su promedio de `mortalidad_infantil`

````python
# Tu código aquí
# 1. Países de América
americas = salud_limpio[salud_limpio["region"] == "Americas"][["pais", "esperanza_vida", "gasto_salud_usd"]]
print(americas)
print("-" * 60)

# 2. Países con esperanza de vida
mayor = salud_limpio[salud_limpio["esperanza_vida"] > 80]
print(f"Cantidad de países: {len(mayor)}")
print(mayor["region"].value_counts())
print("-" * 60)

# 3. Países con niveles bajos y promedio de mortalidad infantil
nivel_ingresos = salud_limpio[salud_limpio["nivel_ingresos"] == "Low"]
print(f"Promedio de mortalidad infantil: {nivel_ingresos['mortalidad_infantil'].mean():.1f}")
print("-" * 60)

# 4. Países con vacunación mayor al 90%
vacunacion = salud_limpio[salud_limpio["vacunacion_pct"] > 90]
print(f"Cantidad de países con vacunación mayor al 90%: {len(vacunacion)}")
print("-" * 60)
````

### 2.5 — Agrupaciones

🔧 **Ejercicio 2.5** — Usa `groupby()` para:
1. Calcular la **esperanza de vida promedio** por `region`
2. Calcular el **gasto en salud promedio** por `nivel_ingresos`
3. Calcular el **número de países** por región
4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?

````python
# Tu código aquí
# 1. Esperanza de vidad promedio por región
#salud_limpio.groupby("region")["esperanza_vida"].mean()
print(salud_limpio.groupby("region")["esperanza_vida"].mean())
print("-" * 60)

# 2. Gasto en salud promedio por nivel de ingresos
print(salud_limpio.groupby("nivel_ingresos")["gasto_salud_usd"].mean())
print("-" * 60)

# 3. Número de países por región
print(salud_limpio.groupby("region")["pais"].count())
print("-" * 60)

# 4. Región con mayor y menor esperanza de vida promedio
esperanza_region = salud_limpio.groupby("region")["esperanza_vida"].mean()
print(f"Región con mayor esperanza de vida: {esperanza_region.idxmax()} - {esperanza_region.max():.1f}")
print(f"Región con menor esperanza de vida: {esperanza_region.idxmin()} - {esperanza_region.min():.1f}")
````

### 2.6 — Nueva columna calculada

🔧 **Ejercicio 2.6**
1. Crea una nueva columna llamada `categoria_vida` aplicando la función `clasificar_esperanza_vida` que creaste en la Parte 1 a la columna `esperanza_vida`
2. ¿Cuántos países caen en cada categoría? (usa `value_counts()`)

````python
# Tu código aquí

# 1. Crear nueva columna
salud_limpio["categoria_vida"] = salud_limpio["esperanza_vida"].apply(clasificar_esperanza_vida)

# 2. Cuántos países por categoría
print(salud_limpio["categoria_vida"].value_counts())
print("-" * 60)

# 3. Porcentaje por categoría
print(salud_limpio["categoria_vida"].value_counts(normalize=True) * 100)
````
