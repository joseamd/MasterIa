Listas

```python
minutos_vistos = [200,350,60,560]
#print(minutos_vistos)

minutos_vistos.append(1050)
print(minutos_vistos)

for min in minutos_vistos:
    print(min)
```

```python
#Total minutos
total_suma = 0

for minutos in minutos_vistos:
    total_suma += minutos

print("El total de minutos es: " , total_suma)
```

```python
#Promedio
promedio = total_suma / len(minutos_vistos)
print("Promedio: ", promedio)
```

```python
#Diccionario
usuarios = [
    {"nombre": "Carolina", "edad": 42, "pais": "Colombia", "minutos":200, "activo": True},
    {"nombre": "Sofia", "edad": 20, "pais": "Brasil", "minutos":350, "activo": True},
    {"nombre": "Adrian", "edad": 30, "pais": "Colombia", "minutos":60, "activo": False},
    {"nombre": "Amada", "edad": 19, "pais": "Polonia", "minutos":560, "activo": True}
]

print(usuarios[0])

for usuario in usuarios:
    print(usuario)    
    

for usuario in usuarios:
    print(usuario["edad"])
```

```python
#Saber los usuarios activos
for us in usuarios:
    if us["activo"] == True:
        print(us["nombre"], "está activo")
```

```python
#Buscar o mostrar en Colombia que hayan visto más de 200 minutos

for usuario in usuarios:
    if usuario["pais"] == "Polonia":
        if usuario["minutos"] > 200:
            print(usuario["nombre"], " es usuario con alto consumo de video")
        
```

```python
#Funciones
def mostrar_activos(usuarios):
    for usuario in usuarios:
        if usuario["activo"]:
            print(usuario["nombre"])
    
```

```python
#Mostrar funcion
mostrar_activos(usuarios)
```
